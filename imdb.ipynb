```python
from bs4 import BeautifulSoup
import requests
import csv
url="https://www.imdb.com/movies-coming-soon/2019-10/"
response=requests.get(url)
soup=BeautifulSoup(response.content,"html.parser")
overview=soup.find_all("img",attrs={"class":"poster shadowed"})
titles=[movie.get("title") for movie in overview]#works fine
#print(titles)
descriptions=soup.find_all("div",attrs={"class":"outline"})#works fine
description=[desc.text.strip(" ") for desc in descriptions]#works fine
#print(description)
scores_divs=soup.find_all("div",attrs={"class":"rating_txt"})
score_div=[div.findChild() for div in scores_divs]
scores=[score.text.strip(" ") if score else "None" for score in score_div ]
#print(scores)
starrcasts=soup.find_all("h5",attrs={"class":"inline"})#works fine
stars=[star for star in starrcasts]#works fine

stars_tags=[i.findNextSiblings("a") for i in stars if i.text== "Stars:"]#works fine

stars_names=[[tag.text for tag in list] for list in stars_tags]
#print(stars_names)
with open('imdb.csv', 'w') as csvfile:
    filewriter =csv.writer(csvfile, delimiter=" ")

    for i in range(len(titles)):
        filewriter.writerow([titles[i],scores[i],description[i],stars_names[i]])
```

```python
from bs4 import BeautifulSoup
import requests
import csv
url="https://www.imdb.com/movies-coming-soon/2019-10/"
response=requests.get(url)
print(response)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

```python
overview=soup.find_all("td",attrs={"class":"overview-top"})
print(overview)
```

```python
results=[]

for card in overview[0:]:
    d={}
    d['titles']=card.find("h4").a.text.strip(" ")
    scores_tag=(card.find("span",attrs={"class":"metascore"}))
    
    if scores_tag:
        d['scores']=scores_tag.text.strip(" ")
    else:
        d['scores']="None"
    d['description']=card.find("div",attrs={"class":"outline"}).text.replace("\n"," ").strip(" ")
    all_star_links=card.find_all("div",attrs={"class":"txt-block"})[1].find_all("a")
    d['stars']=[link.text.replace("\n"," ").strip(" ")for link in all_star_links]
    
    results.append(d)
print(results)


```

```python
with open('imdb_deatils.csv', 'w') as csvfile:
    filewriter =csv.DictWriter(csvfile,fieldnames=['titles','scores','description','stars'])
    filewriter.writerows(results)
```
