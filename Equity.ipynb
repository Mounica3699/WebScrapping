```python
from bs4 import BeautifulSoup
import requests
import csv
url="https://www.bseindia.com/markets/keystatics/Keystat_turnoverequity.aspx"
response=requests.get(url)
print(response)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

```python
table_headings=soup.find("tr",attrs={"class":"tableheading"})
print(table_headings)
```

```python
table_headings_tags=table_headings.find_all("th")
print(table_headings_tags)
```

```python
table_headings_text=[tag.text for tag in table_headings_tags]
print(table_headings_text)
```

```python
table=soup.find("table",attrs={"id":"ContentPlaceHolder1_gvReport"})
table_trs=table.find_all("tr")
print(table_trs)
```

```python
exclude_header_tr=[tr for tr in table_trs[1:]]
print(exclude_header_tr)
```

```python
list_of_trs=[[tr] for tr in exclude_header_tr]
9"tdprint(list_of_trs)
```

```python
list_of_tds=[tr.find_all("td") for trs in list_of_trs for tr in trs]
print(list_of_tds)
```

```python
anchor_tags=soup.find_all("a")
print(anchor_tags)
```

```python
with open('Equity.csv', 'w') as csvfile:
    filewriter =csv.writer(csvfile, delimiter=",")
    filewriter.writerow(table_headings_text)
    for ilist in list_of_tds:
        filewriter.writerow([td.a.text if td.find("a") else td.text for td in ilist])
```
